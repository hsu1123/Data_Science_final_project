```python
camera = 10000
camera_list = []
while len(camera_list) < 40:
    image_url = 'https://cctvn.freeway.gov.tw/abs2mjpg/jpg?camera=' + str(camera)
    res = requests.get(image_url, stream=True)
    file = str(camera)
    if res.status_code == 200:
        if not os.path.isdir(file):
            os.mkdir(file)
        for i in range(5):
            res = requests.get(image_url, stream=True)
            path = file+'/'+ str(i) +'.jpg'
            with open(path,'wb') as file_path:
                for chunck in res:
                    file_path.write(chunck)
            time.sleep(1)
        camera_list.append(camera)
        print("The Image has been downloaded")
    else:
        print("Error!! HTTP Request failed")
    camera += 1
    time.sleep(1)
```
